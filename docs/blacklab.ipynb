```python
from republic_tools.blacklab_client import BlackLabClient

c = BlackLabClient("http://localhost:8080/blacklab-server")
corpus_name='republic-1728'
```

```python
from IPython.display import display, HTML

def display_kwic(patt:str):
    hits = c.get_corpus_hits(corpus_name, patt=patt)["hits"]
    html='<table>'
    for h in hits:
        html += f"<tr><td style='text-align:right;border-style:solid'>{h['start']}</td>"
        for key,align in [("left","right"),("match","center"),("right","left")]:
            html += f'<td style="text-align:{align};border-style:solid">'
            for i,w in enumerate(h[key]["word"]):
                html+=w 
                html += h[key]["punct"][i]
            html += "</td>"
        html += "</tr>"
    html += "</table>"
    display(HTML(html))
```
